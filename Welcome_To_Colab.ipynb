```
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset (pakai pemisah ;)
df = pd.read_csv("Percobaan_mhs.csv", sep=";")

plt.figure(figsize=(8,6))
plt.scatter(
    df["X1"],
    df["X2"],
    s=df["X4"]*10,      # ukuran bubble = X4
    c=df["X3"]          # warna = X3
)

plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Scatter Plot 4 Variabel (X1, X2, X3, X4)")
plt.colorbar(label="X3")
plt.show()
```
